```python
from langchain_openai import ChatOpenAI
import os
from dotenv import load_dotenv
from langchain_core.tools import tool

load_dotenv()
```

```python
llm = ChatOpenAI(model="llama-3.1-8b-instant", 
                 temperature=0.9, 
                 api_key=os.getenv("GROQ_API_KEY"),
                 base_url=os.getenv("GROQ_BASE_URL"))



response = llm.invoke("What is the latest news on the India T20 match?")
print(response.content)
```

```python
from langchain_community.tools import DuckDuckGoSearchRun
from langchain_community.tools import tool
search_tool = DuckDuckGoSearchRun()

@tool("brave_search")
def b_search(query: str) -> str:
    """Searches the web using Brave Search and returns the results."""
    return search_tool.run(query)

print(b_search.name)
print(b_search.description)
print(b_search.args_schema)
```

```python
from langchain.agents import create_agent

agent = create_agent(
     model=llm, 
     tools=[b_search], 
     system_prompt="Answer the question based on the latest news. Use the search tool to find the latest news if needed."
)

reponse = agent.invoke(
    {"messages": [{"role": "user", "content": "Give me the latest points table in T20 world cup 2026?"}]}
)

print(reponse["messages"][-1].content)
```
